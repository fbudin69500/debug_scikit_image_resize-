```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
#Some nice default configuration for plots
plt.rcParams['figure.figsize'] = 30, 30
plt.gray()
```

```python
l=np.load('resized.npy')
t=np.load('resized_2.npy')
```

```python
plt.figure()
plt.imshow(l)

list_points={}
print l.shape
print t.shape
plt.figure()
for i in range(0,l.shape[0]):
    for j in range(0,l.shape[1]):
        if abs(l[i,j] - t[i,j]) > 0.000001:
            plt.plot(i, j, 'r+')
            if j in list_points:
                list_points[j]=list_points[j]+[i]
            else:
                list_points[j]=[i]

plt.imshow(t)
```
